## **Challenge 1**

#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has **450 seats**. 
- Based on historical data we conclude that **each individual passenger has a 3% chance of missing it's flight**. 

If the Ironhack Airlines **routinely sells 460 tickets**, `what is the chance that they have a seats for all passenger?`

```python
# plane_seats = 450
# p(missed flight) = 0.03

# ironhack_airlines sells 460 tickets
# What is the probability that more passengers show up than there are seats on the plane?
import numpy as np
from scipy.stats import binom
plane_seats = 450
p_missed_flight = 0.03
tickets_sold = 460
# The number of passengers that show up follows a binomial distribution with parameters n = 460 and p = 0.97 (since 3% miss the flight).
n = tickets_sold
p_show_up = 1 - p_missed_flight
# We want to find the probability that more than 450 passengers show up, which is equivalent
# to finding the probability that 451 or more passengers show up.
# This can be calculated as 1 - P(X <= 450), where X is the number of passengers that show up.
# Calculate P(X <= 450) using the cumulative distribution function (CDF) of the binomial distribution.
p_at_most_450 = binom.cdf(plane_seats, n, p_show_up)
# Now calculate the probability that more than 450 passengers show up.
p_more_than_450 = 1 - p_at_most_450
print(f"The probability that more passengers show up than there are seats on the plane is: {p_more_than_450:.4f}")
```
